```python
from scipy.stats import spearmanr

worldpop2013Column = poplife_merge['POPULATION']
lifeResetColumn = poplife_merge['SP.DYN.LE00.IN']
(correlation, pValue) = spearmanr(worldpop2013Column, lifeResetColumn)
print('The correlation is', correlation)
if pValue < 0.05:
    print('It is statistically significant.')
else:
    print('It is not statistically significant.')
```

```python
#Comparison between top 10 countries with highest GDP and countries with the longest life expectancy,

gdpclean= gdpReset.rename(columns={'NY.GDP.MKTP.CD' : 'GDP'})
gdpclean[34:]
lifeclean= lifeReset.rename(columns={'SP.DYN.LE00.IN'  : 'LIFE EXPECTANCY'})
lifeclean[34:]
gdplife_merge = pd.merge(gdpclean, lifeclean).dropna()
gdplife_merge[34:]
```

```python
from scipy.stats import spearmanr

gdpColumn = gdplife_merge['GDP']
lifeColumn = gdplife_merge['LIFE EXPECTANCY']
(correlation, pValue) = spearmanr(gdpColumn, lifeColumn)
print('The correlation is', correlation)
if pValue < 0.05:
    print('It is statistically significant.')
else:
    print('It is not statistically significant.')
```

```python
gdplife_merge.sort_values(by = 'GDP', ascending = False).head(10)
```
